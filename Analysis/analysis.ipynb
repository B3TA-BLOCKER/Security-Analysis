**Importing the required modules**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**The Variable Survey is having the Dataset**

```python
Survey = pd.read_csv('Cyber Risk Assessment.csv')
```

**A concise summary of the DataSet**

```python
Survey.info()
```

```python
Survey = Survey.dropna(subset=['Write the name of your Enterprise.', 'Write your current role at the Enterprise ?'])

# Creating the bar plot
sns.barplot(x='Write the name of your Enterprise.', y='Write your current role at the Enterprise ?', data=Survey)

# Set labels for x and y axes
plt.xlabel('Enterprise')
plt.ylabel('Role')

# Rotating x-axis labels
plt.xticks(rotation=90)

plt.show()
```

**Enterprises using licensed Software**

```python
# Filter the DataSet to include only rows where the answer is "Yes"
yes_license_df = Survey[Survey['   Do all applications and/or software used in your organization have current licenses?  '] == 'Yes']

yes_license_df = yes_license_df.dropna(subset=['Write the name of your Enterprise.', 'Write your current role at the Enterprise ?'])

# Create the count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='Write the name of your Enterprise.', data=yes_license_df, hue='Write your current role at the Enterprise ?')

# Set labels for x and y axes
plt.xlabel('Enterprise')
plt.ylabel('')

# Rotate x-axis labels
plt.xticks(rotation=90)

# Show the plot
plt.legend(title='Role', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
target_column = '  What number of citizens would this business failure impact?  '

# Drop rows with missing values in the selected column
Survey_cleaned = Survey.dropna(subset=[target_column])

# Count the frequency of each response
response_counts = Survey_cleaned[target_column].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
response_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Impact on Citizens')
plt.ylabel('')  # Remove the y-axis label
plt.show()
```

```python
# Select the column containing the responses to the question
target_column = '  Have you developed backup strategies for continuity of automated services should you have an IT failure?    '


# Filter the data for "Yes" and "No" responses
yes_responses = Survey[Survey[target_column].str.strip().str.lower() == 'yes']
no_responses = Survey[Survey[target_column].str.strip().str.lower() == 'no']

# Count the frequency of each response
yes_count = yes_responses.shape[0]
no_count = no_responses.shape[0]

# Plot the bar graph
plt.figure(figsize=(8, 6))
plt.bar(['Yes', 'No'], [yes_count, no_count], color=['blue', 'red'])
plt.title('Backup Strategies for IT Failure Continuity')
plt.xlabel('Response')
plt.ylabel('Count')
plt.show()
```

```python
# Define the target column
target_column = '  Are information security risks compared to the established risk criteria and prioritised?   '

# Filter the data for "Yes" and "No" responses
yes_responses = Survey[Survey[target_column].str.strip().str.lower() == 'yes']
no_responses = Survey[Survey[target_column].str.strip().str.lower() == 'no']

# Count the frequency of each response
yes_count = len(yes_responses)
no_count = len(no_responses)

# Plot the bar graph
plt.figure(figsize=(8, 6))
plt.bar(['Yes', 'No'], [yes_count, no_count], color=['green', 'red'])
plt.title('Comparison and Prioritization of Information Security Risks')
plt.xlabel('Response')
plt.ylabel('Number of Enterprises')
plt.show()
```

```python
# Define the target column
target_column = 'If your organization uses local IDS, do you use “host-based” IDS (HIDS) or “network-based” IDS (NIDS) or a combination of both?'

# Count the frequency of each response
response_counts = Survey[target_column].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(response_counts, labels=response_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Usage of Local IDS')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
